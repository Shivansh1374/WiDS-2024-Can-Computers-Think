```python
#Below is the dictionary for frozen lake
```

```python
fl_mdp = {}
```

```python
for state in range(0, 16):

    transitions = {}

    for action in ["Up", "Down", "Right", "Left"]:
        transitions[action]=[]
        if(state==5 or state==7 or state==11 or state==12 or state==15):
            transitions[action]=[(1, state, 0, True)]
        else:
            if(action!="Up"):
                if(state==13 or state==14):
                    transitions[action].append((1/3, state, 0, False))
                elif(state==1 or state==3 or state==8):
                    transitions[action].append((1/3, state+4, 0, True))
                else:
                    transitions[action].append((1/3, state+4, 0, False))
            if(action!="Down"):
                if(state==0 or state==1 or state==2 or state==3):
                    transitions[action].append((1/3, state, 0, False))
                elif(state==9):
                    transitions[action].append((1/3, state-4, 0, True))
                else:
                    transitions[action].append((1/3, state-4, 0, False))
            if(action!="Right"):
                if(state==0 or state==4 or state==8):
                    transitions[action].append((1/3, state, 0, False))
                elif(state==6 or state==13):
                    transitions[action].append((1/3, state-1, 0, True))
                else:
                    transitions[action].append((1/3, state-1, 0, False))
            if(action!="Left"):
                if(state==3):
                    transitions[action].append((1/3, state, 0, False))
                elif(state==4 or state==6 or state==10):
                    transitions[action].append((1/3, state+1, 0, True))
                elif(state==14):
                    transitions[action].append((1/3, state+1, 1, True))
                else:
                    transitions[action].append((1/3, state+1, 0, False))
                    

    fl_mdp[state] = transitions
```
